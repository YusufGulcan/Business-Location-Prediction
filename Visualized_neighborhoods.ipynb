```python
import pandas as pd 
import numpy as np 
import json 
import requests
import geopandas as gpd
from  google.colab import drive
import folium
```

```python
driver = drive.mount('/content/drive/')
```

```python
path1 = '/content/drive/My Drive/Database/predicted.csv'                      ## Data frame created and saved in the previous step
data1 = pd.read_csv(path1)


url = 'https://ckan0.cf.opendata.inter.prod-toronto.ca/dataset/neighbourhoods/resource/1d38e8b7-65a8-4dd0-88b0-ad2ce938126e/download/Neighbourhoods.geojson'              ## This is the url for the geojson file. 

re = requests.get(url).text                         ## Create a request order for geojson file and convert it to text format.
loca=gpd.read_file(re)                                ## read the file via gpd and it becomes a geopandas data frame.
loca.head(3)
```

**The cell below is provided in the folium documentation, I only modify the parts for my own data.**

 **Opportunity Scale**

```python
m = folium.Map(location=[43.80, -79.2577140359756], zoom_start=11)                       ## creates a map object and zooms in the specified location.

folium.Choropleth(                                                                  ## Creates the filled map layer on top of the classic map
    geo_data=loca,                                                                 ## geojson data is specified. This should be a file with a geometry data.
    name="choropleth",                                                             ## did not change it
    data=data1,
    columns=["Unnamed: 0", "preds"],                                                ## column names in the data frame I created in the previous step.
    key_on="feature.properties.AREA_NAME",                ## This is the tricky part. You need to check the geojson file and look at the directory of the geometry column. It matches the data from two sources like that.
    fill_color="YlGn",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="preds",                                 ## The colors are different for each unique value in this column. Our 'preds' column includes the clusters, so each cluster is painted in different color.
).add_to(m)                                              ## Add the filled map layer to the map object
folium.TileLayer('cartodbpositron').add_to(m)
                      

m
```

**Crime Rates Per Neighborhood**
